```python
def validacion_nombre (respuestas_candidatos):
  """
  Función para validar que el nombre contenga nombre y apellidos, sin números ni simbolos.

  Argumeno recibido: Diccionario con la información del candidato.
  Retorna el mismo diccionario validado y modificado de ser necesario.
  """
  nombre_comprobar = respuestas_candidatos["nombre"]
  while True:
    try:
      if len(nombre_comprobar.split()) < 2:
        raise ValueError(
            f"El nombre introducido {nombre_comprobar} no es valido.\n"
            "Debe facilitar un nombre y apellido.\n"
        )
      if any(letra.isdigit() for letra in nombre_comprobar):
        raise ValueError(
            f"El nombre introducido {nombre_comprobar} no es valido."
            "Su nombre no puede contener números.\n"
        )
      respuestas_candidatos["nombre"] = nombre_comprobar
      return respuestas_candidatos
    except ValueError as e:
      print(f"\nError: {e}")
      nombre_comprobar = input("Por favor, introduzca su nombre y apellido nuevamente:\n")
```

```python
def validacion_telefono (respuestas_candidatos):
  """
  Función para validar que el número de telefono sea correcto.

  Argumeno recibido: Diccionario con la información del candidato.
  Retorna el mismo diccionario validado y modificado de ser necesario.
  """
  telefono = respuestas_candidatos["telefono"]
  while True:
    try:
      telefono = str(telefono).replace(" ", "").replace("-", "").replace(".","").replace(",", "")
      if not telefono.isdigit():
        raise ValueError (
            f"Su telefono número: {telefono}, no es valido"
            "debe contener solo digitos\n"
        )
      if len(telefono) < 9 or len(telefono) > 12:
        raise ValueError(
            f"Su telefono número: {telefono} debe tener entre 9 y 12 digitos\n"
            )
      respuestas_candidatos["telefono"] = telefono
      return respuestas_candidatos
    except ValueError as e:
      print(f"\nError: {e}")
      telefono = input("Por favor, introduzca su numero de telefono nuevamente:\n")
```

```python
def validacion_experiencia (respuestas_candidatos):
  """
  Función para validar que el número de años trabajados en la empresa
  sea un número entero.

  Argumeno recibido: Diccionario con la información del candidato.
  Retorna el mismo diccionario validado y modificado de ser necesario.
  """
  experiencia = respuestas_candidatos["experiencia"]
  while True:
    try:
      if isinstance(experiencia, str):
        try:
          experiencia = float(experiencia)
        except ValueError:
          raise ValueError(
              f"No se permite texto en el campo Experiencia.\n"
              "Debe facilitar un número entero redondeado de años trabajados en la empresa.\n"
          )
      if isinstance(experiencia, float) and not experiencia.is_integer():
        raise ValueError(
            f"No se permiten números con decimales en el campo Experiencia como {experiencia}.\n"
            "Debe facilitar un número entero redondeado de años trabajados en la empresa.\n"
        )
      experiencia = int(experiencia)
      if experiencia < 0:
        raise ValueError(
            "Debe presentar años positivos trabajados en la empresa.\n"
            )
      respuestas_candidatos["experiencia"] = experiencia
      return respuestas_candidatos
    except ValueError as e:
      print(f"\nError: {e}")
      experiencia = input("Por favor, introduzca el número de nuevo:\n ")
```

```python
def validacion_niveles(respuestas_candidatos):
  """
  Funcion para validar que los datos referentes a la experiencia en los campos:
  Catalán, Excel y titulacion academica esten dentro de los parámetros.

  Argumeno recibido: Diccionario con la información del candidato.
  Retorna el mismo diccionario validado y modificado de ser necesario.
   """
  for tema, nota in [("catalan", respuestas_candidatos["catalan"]),
                     ("excel", respuestas_candidatos["excel"]),
                     ("titulo", respuestas_candidatos["titulo"])]:
    while True:
      try:
        if not nota.isdigit():
          raise ValueError(
              f" El valor: {nota}, en el campo {tema} no es valido.\n"
              "No se permite texto ni simbolos. Debe introducir un número entre el 0 y el 3 .\n"
              )
        nota = int(nota)
        if nota not in [0, 1, 2, 3]:
          raise ValueError(
              f"El valor: {nota}, en el campo {tema} está fuera de parametros.\n"
              "Debe introducir un número entre el 0 y el 3.\n"
              )
        respuestas_candidatos[tema] = nota
        break
      except ValueError as e:
        print(f"\nError: {e}")
        nota = input(f"Por favor, introduzca su nivel de {tema} nuevamente:\n")

  return respuestas_candidatos
```

```python
def validaciones (respuestas_candidatos):
  """
  Funcion que unifica las validaciones de los campos solicitados por RRHH.

  Argumeno recibido: Diccionario con la información del candidato.
  Retorna el mismo diccionario validado y modificado de ser necesario.
  """
  validacion_nombre(respuestas_candidatos)
  validacion_telefono(respuestas_candidatos)
  validacion_experiencia(respuestas_candidatos)
  validacion_niveles(respuestas_candidatos)

  return respuestas_candidatos
```

```python
def calcular_puntuacion (respuestas_candidatos):
  """
  Funcion para calcular la puntuacion de cada candidato

  Argumento recibido: Diccionario con la información del candidato.
  Retorna el mismo diccionario con nueva clave "puntuacion y su valor calculado.
  """
  puntuaciones = (respuestas_candidatos["experiencia"] * 1.5 +
                respuestas_candidatos["catalan"] +
                respuestas_candidatos["excel"] +
                respuestas_candidatos["titulo"])
  respuestas_candidatos["puntuacion"] = puntuaciones

  return respuestas_candidatos
```

```python
def clasificacion_candidato (respuestas_candidatos):
  """
  Función para clasificar a cada candidato según su puntuación.

  Argumento recibido: Diccionario con la información del candidato.
  Retorna el mismo diccionario con dos nuevas claves "clasificacion" y
  "posición clasificación" y sus valores.
  """
  clasificacion = ""
  posicion_clasificacion = 0
  if respuestas_candidatos["puntuacion"] >=  12:
      clasificacion = "Prioridad Alta"
      posicion_clasificacion = 1
  elif respuestas_candidatos["puntuacion"] >= 8:
      clasificacion = "Candidato Adecuado"
      posicion_clasificacion = 2
  else:
      clasificacion = "No prioritario"
      posicion_clasificacion = 3
  respuestas_candidatos["clasificacion"] = clasificacion
  respuestas_candidatos["posicion_clasificacion"] = posicion_clasificacion

  return respuestas_candidatos
```

```python
def solicitud_candidato(contador, respuestas_candidatos):
    """
    Funcion que recoge las respuestas del candidato y las muestra en pantalla.

    Argumentos recibidos:
      1. Contador para establecer nº de solicitud.
      2. Dicionario con la información del candidato.
    No retorna nada ya que lo imprime en pantalla.
    """
    print(f"\nHa realizado la solicitd nº {contador} con los siguintes datos:\n"+
        f"Nombre y apellido: {respuestas_candidatos['nombre']}.\n"+
        f"Teléfono: {respuestas_candidatos['telefono']}.\n"+
        f"Años trabajados en la empresa: {respuestas_candidatos['experiencia']} años.\n"+
        f"Nivel de conocimiento en Catalán: {respuestas_candidatos['catalan']}.\n"+
        f"Nivel de conocimiento en Excel: {respuestas_candidatos['excel']}.\n"+
        f"Ha obtenido una puntuación final de: {respuestas_candidatos['puntuacion']} puntos.\n\n"+
        "Se le contactara proximamente desde el departamento de Recursos Humanos.\n\n")
```

```python
def informa_rrhh(candidatos):
  """
  Función que prepara el informe solicitado por RRHH, ordena los candidatos por
  la posición en la clasificación y, de haber empate, por su puntuación.

  Argumento recibido con la lista de diccionarios de los 10 candidatos aceptados.
  No devuelve nada, imprime toda la información.
  """
  resultados_ordenados = sorted(
      candidatos, key = lambda x: (x["posicion_clasificacion"], -x["puntuacion"])
      )
  print("Informe de candidatos:\n")
  for candidato in resultados_ordenados:
    print(f"Nombre y apellido: {candidato['nombre']}.\n"+
          f"Teléfono: {candidato['telefono']}.\n"+
          f"Puntuación final: {candidato['puntuacion']}.\n\n")
```

```python
def proceso_seleccion():
  """
  Función que dirige las acciones según solicitudes recibidas.

  No recibe ni retorna nada.
  """
  lista_candidatos = []
  contador = 1
  while True:
    if contador <= 10 :
        respuestas_candidatos = datos_candidato() #Se obtienen info de los candidatos
        respuestas_validadas = validaciones(respuestas_candidatos) #Se valida info
        resultados_calculados = calcular_puntuacion(respuestas_validadas) #Se calcula puntuacion
        resultados_clasificados = clasificacion_candidato (resultados_calculados)  #se hace la clasificacion del candidato
        solicitud_candidato(contador, resultados_clasificados) #informe para el candidato
        lista_candidatos.append(resultados_clasificados)
        contador += 1
    else:
        informa_rrhh(lista_candidatos)
        print("Ya no se aceptan más solicitudes")
        break


```

```python
def procesar_clientes(lista_clientes):
  """
  Función que pasa la lista de diccionarios, con la info de los clientes a un
  diccionario de diccionarios, donde la clave principal es el id_cliente y su
  valor el diccionario con el resto de información. Así mismo, la función procesa
  tres funciones más para ampliar la información del cliente.

  Argumento recibido: lista de diccionarios con la información de los clientes.
  Retorna un diccionario de diccionarios con la información de los clientes
  ampliada.
  """
  diccionario_clientes = {
    cliente["id_client"]: {k: v for k, v in cliente.items() if k != "id_client"}
    for cliente in lista_clientes
    }

  for datos_cliente in diccionario_clientes.values():

    datos_cliente.update(obtener_segmento_y_factor(datos_cliente["codi_comportament"])) #añade elementos al diccionario según el valor del código de comportamiento.
    datos_cliente["f_edad"] = obtener_factor_edad(datos_cliente["grup_edat"]) #Crea un nuevo elemento
    datos_cliente["riesgo_final"] = calcular_riesgo(datos_cliente["f_codigo"], datos_cliente["f_edad"])

  return diccionario_clientes
```

```python
def convertir_lista_dic(lista_pacientes):
  """
  Función que tomas los datos de la lista facilitada y la convierte
  en una lista de diccionarios, donde la información de cada paciente
  esta en un diccionario.

  Arguemto reibido: Lista de pacientes
  Retorna una lista de diccionarios
  """
  lista_separada = []
  lista_claves = ["nombre", "edad", "sintomas"]
  for paciente in lista_pacientes:
    paciente_separado = paciente.split("|") #separa la informacion de cada elemento de la lista por donde hay |
    lista_separada.append(paciente_separado) #añade cada elemento a una nueva lista

  lista_dic_pacientes = []
  for paciente in lista_separada:
    diccionario_paciente = dict(zip(lista_claves, paciente)) #crea diccionario en cada elemento de la lista
    lista_dic_pacientes.append(diccionario_paciente)#convirtiendose en una lista de

  return lista_dic_pacientes
```

```python
#Prueba función:
l_dic_prueba = convertir_lista_dic(lista_pacientes)
```

```python
def normalizacion_sintomas(lista_dic_pacientes):
  """
  Función que normaliza los valores de los sintomas
  todo en minusculas, separando cada sintoma por donde esta la ",",
  limpiando los espacios en blanco al inicio y final y uniendo las palabras con _.

  Argumento recibido: lista de diccionarios
  Retorna lista de diccionarios
  """
  for paciente in lista_dic_pacientes:
    paciente["sintomas"] = paciente["sintomas"].lower() # paso a minusculas los sintomas
    paciente["sintomas"] = paciente["sintomas"].split(",") # separa los distintos sintomas por donde esta la coma
    paciente["sintomas"] = [sintoma.strip() for sintoma in paciente["sintomas"]] #limpieza espacios en blanco
    paciente["sintomas"] = [sintoma.replace(" ", "_") for sintoma in paciente["sintomas"]] #cambio espcaio en blanco intermedio por "_".

  return lista_dic_pacientes
```

```python
def asignacion_riesgo(lista_pacientes):
  """
  Función que calcula el riego del paciente, en función de los sintomas
  presentados y lo añade a la clave "riesgo" al diccionario según las
  puntuaciones por sintoma facilitadas por el hospital.

  Argumento recibido: Lista diccionario
  Restorna la misma lista de diccionarios, con los diccionarios ampliados
  """
  dic_nivel_riesgo = {
    "fiebre_alta": 3,
    "dificultad_respirar": 5,
    "dolor_pecho": 4,
    "tos": 1,
    "fatiga": 1,
    "dolor_cabeza": 1
} #Se deja el diccionario como variable local de la función para que no se vaya
#arrastrando desde el inicio y de necesitar cambiar valores solo cambiarlos aqui.

  for paciente in lista_pacientes:
    sintomas = paciente["sintomas"] #bucle para iterar hasta los sintomas
    suma_sintomas = 0
    for sintoma in sintomas: #bucle para iterar dentro de la lista de sintomas
      if sintoma in dic_nivel_riesgo:
        suma_sintomas += dic_nivel_riesgo[sintoma]
    paciente["riesgo"] = suma_sintomas

  return lista_pacientes
```

```python
def obtener_df_pacientes(lista_diccionario_con_riesgo):
  """
  Pasa el diccionario final a un df
  y se devuelve ya ordenado de mayor a menor riesgo.

  Argumento recibido: Lista de diccionarios
  Retorna un DataFrame
  """

  #Conversión de la lista de diccionarios en df:
  df_hospital = pd.DataFrame(lista_diccionario_con_riesgo)
  #Ordena el df según la columna de riesgo de mayor a menor:
  df_hosp_ordenado = df_hospital.sort_values(by = ["riesgo"], ascending = False)
  #Inicia lo el indice en 1 y termina en la longitud del df+1:
  df_hosp_ordenado.index = pd.RangeIndex(start = 1, stop = len(df_hospital) + 1)
  #En cada linea del df, en la columna sintomas pasa convierte en string cada elemento y los une mediante una ",":
  df_hosp_ordenado["sintomas"] = [', '.join(map(str, x)) for x in df_hosp_ordenado["sintomas"]]


  return df_hosp_ordenado
```

```python
def gestion_pacientes_urgencia (lista):
  """
  Función que ejecuta y organiza la atención de los pacientes
  según la lista facilitada de pacientes.

  Argumento recibido: Lista con la información de clientes.
  Retorna el df ordenado
  """
  lista_diccionario_pacientes = convertir_lista_dic(lista)
  lista_dic_pacientes_sintomas_normalizados = normalizacion_sintomas(lista_diccionario_pacientes)
  lista_final = asignacion_riesgo(lista_dic_pacientes_sintomas_normalizados)

  return obtener_df_pacientes(lista_final)
```

La función principal (calculo_precios_seguros) unifica una serie de funciones con tal de conseguir ampliar la información de los clientes (info guardad en lista de diccionarios) con la información de riesgo de fraude (info en diccionario de diccionarios), calcular precio seguro final y guardar esta información en un DataFrame y resolver preguntas realizadas por el departamento.

Subfunciones:
1.   Función unir info clientes seguro: Amplia la informacion de cada cliente con la información que le pertoca de nivel de fraude según sel valor del codigo establecido en los datos del cliente.
2.   Función calculo precio final: Calcula el precio final de la poliza según los datos del cliente.
3. Función creación resultado df: Crea un dataframe segun los datos obtenidos. Renombra el nombre de las columnas y ordena los clientes según el valor del fraude de mayor a menor y según el precio final de menor a mayor.
4. Función análisi datos: Funcion que recoge las preguntas, calcula las preguntas sobre el DataFrame y muestra las respuestas.

```python
import pandas as pd
import copy
```

```python
clients = [
    {"nom": "Ana", "edat": 42, "codi": 0},
    {"nom": "Carlos", "edat": 30, "codi": 2},
    {"nom": "Isabel", "edat": 55, "codi": 3},
    {"nom": "Jorge", "edat": 40, "codi": 1},
    {"nom": "Marta", "edat": 28, "codi": 0}
]

nivells_seguro = {
    0: {"categoria": "Baix risc", "preu_base": 120, "fraude": 0.01},
    1: {"categoria": "Risc mitjà", "preu_base": 200, "fraude": 0.05},
    2: {"categoria": "Risc alt", "preu_base": 350, "fraude": 0.15},
    3: {"categoria": "Risc crític", "preu_base": 500, "fraude": 0.30}
}
```

```python
def unir_info_clientes_seguro(lista_diccionarios, diccionario_diccionarios):
  """
  Amplia la informacion de cada cliente con la información que le pertoca
  de nivel de fraude según sel valor del codigo establecido en los datos del cliente.

  Argumento recibidos:
    1. Lista de diccionarios con información de clientes.
    2. Diccionario de diccionarios con la información de las categorias riesgos.
  Retorna la lista de diccionarios con la info de los clientes ampliada y los valores borrados.
  """
  #Hago deep copies para que se guarden los datos de los diccionarios anidados
  lista_clientes = copy.deepcopy(lista_diccionarios)
  dic_niveles = copy.deepcopy(diccionario_diccionarios)
  valores_borrados = {}
  for cliente in lista_clientes:
    valor_codigo = cliente["codi"]#Para saber que valor tiene el codigo con el que hay que ampliar la información de cliente
    cliente.update(dic_niveles[valor_codigo]) #Amplia el diccionario de cada cliente con los elementos que tiene el diccionario de niveles
    valores_borrados[cliente["nom"]] = cliente.pop("codi")#Por tal de no tener valores redundantes, elimino el elemento del codigo de cada cliente
    #y por si acaso los guardo junto con el nombre del cliente (k:v)

  return lista_clientes, valores_borrados
```

```python
#Prueba función:
ld_a = unir_info_clientes_seguro(clients, nivells_seguro)
```

```python
def calculo_precio_final(lista_clientes):
  """
  Calcula el precio final de la poliza según los datos del cliente.

  Argumento recibido: lista de diccionarios con la info de clientes.
  Retorna la misma lista con nuevo elemento "precio_final".
  """
  for cliente in lista_clientes:
    cliente["precio_final"] = round(cliente["preu_base"]*( 1 + cliente["edat"] / 100), 2)

  return lista_clientes
```

```python
#Prueba función:
ld_b = calculo_precio_final(ld_a)
```

```python
def creacion_resultados_df (lista_diccionarios):
  """
  Crea un dataframe segun los datos obtenidos.
  Renombra el nombre de las columnas y ordena los clientes según el valor del
  fraude de mayor a menor y según el precio final de menor a mayor.

  Argumento recibido: Lista de diccionarios con datos de cliente.
  Retorna DataFrame y lo muestra en pantalla.
  """
  df_seguro = pd.DataFrame(lista_diccionarios).rename(columns= {
    "nom":"nombre",
    "edat":"edad",
    "categoria": "Categoria",
    "preu_base": "precio_base",
    "fraude": "fraude",
    "precio_final": "precio_final"
    }).sort_values(by = ["fraude",
                     "precio_final"], ascending = [False, True])#Se actualiza el indice.

  df_seguro.index = range(1, len(df_seguro) + 1)#Se actualiza el indice, iniciando en valor 1.
  display(df_seguro)
  return df_seguro
```

```python
#Prueba función:
creacion_resultados_df(ld_b)
```

```python
def analisis_datos(df_seguro):
  """
  Funcion que recoge las preguntas, calcula las preguntas sobre el DataFrame
  y muestra las respuestas.

  Argumento recibido: Dataframe
  No retorna nada, muestra en pantalla
  """

  #Se buca el valor máximo de la columna de fraude
  #y se muestra la información del cliente o clientes cuyo valor de fraude sea
  #igual al máximo.
  print("\n\n1.Que cliente tiene más probabilidades de fraude?")
  valor_fraude_max = df_seguro["fraude"].max()
  df_pregunta_1 = df_seguro[df_seguro["fraude"] == valor_fraude_max]
  print("\nLos datos del cliente o clientes con mayor probabilidad de fraude son:\n")
  display(df_pregunta_1)

  #Se buca el valor máximo de la columna de precio final
  #y se muestra la información del cliente o clientes cuyo valor de precio final
  #sea igual al máximo.
  print("\n\n2.Quien tiene el precio final más elevado?")
  valor_precio_final_max = df_seguro["precio_final"].max()
  df_pregunta_2 = df_seguro[df_seguro["precio_final"] == valor_precio_final_max]
  print("\nLos datos del cliente o clientes con mayor precio final son:\n")
  display(df_pregunta_2)

  #Se calcula nuevo campo "puntuacion_riesgo"
  #Se buca el valor máximo de la columna puntuacion riesgo
  #y se muestra la información del cliente o clientes cuyo valor de precio final
  #sea igual al máximo.
  print("\n\n3.Quien seria prioritario de revisar manualmente?")
  df = df_seguro.copy()
  df["puntuacion_riesgo"] = df["fraude"] * df["precio_final"]
  valor_riesgo_max = df["puntuacion_riesgo"].max()
  df_pregunta_3 = df[df["puntuacion_riesgo"] == valor_riesgo_max]
  print("\nLos datos del cliente o clientes que debaria ser reivisado manualmente:\n")
  display(df_pregunta_3)
```

```python
def calculo_precios_seguro(lista, diccionario):
  """
  Función general que orquesta todas las funciones.

  Argumento recibidos:
  1. Lista de diccionarios con información de clientes.
  2. Diccionario de diccionarios con la información de las categorias riesgos.
  No retorna nada, se muestran los resultados de la función análisis datos.
  """
  info_unida, valores_codigo_borrados = unir_info_clientes_seguro(lista, diccionario)
  info_completa = calculo_precio_final(info_unida)
  df_final = creacion_resultados_df(info_completa)
  analisis_datos(df_final)
```

```python
def definir_longitud():
  """
  Función donde se solicita la longitud de la contraseña, número entero entre
  el 6 y el 12 y se comprueba que este dentro de los valores.

  No recibe argumentos.
  Retorna número entero en la variable longitud
  """
  print(
      'Bienvenido a la aplicación de creación de contraseñas.\n'
      'La contraseña debe tener, como mínimo, un caracter en mayusculas, minuscula, número y caracter especial.'
      '\nPasaremos a determinar la longitud de la contraseña, ha de ser como mínimo de 6 caracteres.\n')

  longitud_str = input("Quiere una longitud determinada de su nueva contraseña?\n")
  while True:
    try:
      if not longitud_str.isdigit():
        raise ValueError(
            f"Error en formato: El valor '{longitud_str}' no es válido, debe de ser un número entero."
        )

      longitud = int(longitud_str)

      if longitud < 6:
        raise ValueError(
            f"Error en la longitd: La contraseña no puede tener {longitud} caracteres.\nDebe tener entre 6 y 12 caracteres"
        )
      if longitud > 12:

        raise ValueError(
            f"Error en la longitd: La contraseña no puede tener {longitud} caracteres.\nDebe tener entre 6 y 12 caracteres"
        )
      print(f"\nLa longitud de su contraseña será de {longitud} caracteres.\n")
      break

    except ValueError as e:
      print(f"\n{e}")
      longitud_str = input("\nPor favor, facilite un número entero entre 6 y 12 para establecer la longitud de su contraseña.\n")

  return longitud
```

```python
def definir_caracteres():
  """
  Función que solicita si se quieren carateres en mayuscula, minuscula, números
  y carácteres especiales mediante la función definir boolenado.
  Obliga que como se quiera un tipo de caracter.

  No recibe argumentos
  Retorna valores booleanos para las variables:
  mayusculas, minusculas, numeros, caracteres_especiales
  """

  while True:

    mayusculas = definir_booleano("mayusculas")
    minusculas = definir_booleano("minusculas")
    numeros = definir_booleano("numeros")
    caracteres_especiales = definir_booleano("caracteres especiales")

    if any([mayusculas, minusculas, numeros, caracteres_especiales]):
        break
    print("\nDebe seleccionar al menos un tipo de carácter.\n")


  return mayusculas, minusculas, numeros, caracteres_especiales
```

```python
def definir_booleano(categoria):
  """
  Gestiona la respuesta de si se quiere o no los caracteres especiales

  Argumento recibido: Categoría de tipo de dato a definir en la contraseña.
  Retorna el valor booleano.
  """

  while True:
      respuesta = input(
      f"Quiere que la contraseña tenga {categoria}? (s/n).\n"
      ).strip().lower() #elimina espacios en blanco, covierte en minusculas el valor del input
      if respuesta in ("s", "n"):
        return respuesta == "s"
      print(f"\nValor introducido '{respuesta}' no valido.\n")#else
```

```python
def generar_contrasenya (longitud=12, mayusculas=True, minusculas=True, numeros=True, signos=False):
  """
  Función que genera una constraseña aleatoria con mayusculas, minusculas,
  números y caracteres especiales segun una longitud y cantidad de caracteres
  dada por el usuario. Se comprueba que como mínimo hay un caracter de cada y
  devuelve la contraseña al usuario.

  Argumentos recibidos:
    1. Valor número entero de la longitud de la contraseña.
    2. Valores booleanos de mayusculas, minusculas, numeros, caracteres_especiales
  Retorna los caracter generados aleatoriamente unidos en una variable (contraseña).
  """

  caracteres = ""
  caracteres_obligatorios = []
  if mayusculas:
    caracteres += string.ascii_uppercase
    caracteres_obligatorios += ''.join(np.random.choice(list(string.ascii_uppercase)))
  if minusculas:
    caracteres += string.ascii_lowercase
    caracteres_obligatorios += ''.join(np.random.choice(list(string.ascii_lowercase)))
  if numeros:
    caracteres += string.digits
    caracteres_obligatorios += ''.join(np.random.choice(list(string.digits)))
  if signos:
    caracteres += string.punctuation
    caracteres_obligatorios += ''.join(np.random.choice(list(string.punctuation)))

  contrasenya = [np.random.choice(list(caracteres)) for i in (range(longitud - len(caracteres_obligatorios)))]

  contrasenya_final = caracteres_obligatorios + contrasenya
  np.random.shuffle(contrasenya_final)

  return "".join(contrasenya_final)
```

```python
def nueva_contrasenya ():
  """
  Función que recoge el resto de funciones y devuelve la nueva contraseña.

  No recibe argumentos.
  Retorna la contraseña
  """

  longitud = definir_longitud()
  mayus, minus, num, sig = definir_caracteres()
  contra = generar_contrasenya(longitud, mayus, minus, num, sig)

  return contra
```

```python
def preparacion_archivo (ruta):
  """
  Función que se encarga de abrir el archivo de texto,
  limpiarlo y dejarlo en formato lista  para que se pueda convertir en DF.

  Argumento recibido: Variable con la ruta del archivo que se tiene que leer.
  Retorna lista de listas generada.
  """

  with open(ruta, 'rt', encoding='utf-8') as fichero:#se abre fichero
    lista_fichero = fichero.readlines()#se lee como lista

  lista_partidos =[]

  for elemento in lista_fichero: #se trata cada elemento de la lista
    partido = elemento.replace('\n', '').replace('-','\t').split('\t')#elimino \n, sustituyo el guión de los puntos por \t y se separan por\t
    partido[1] = int(partido[1])
    partido[2] = int(partido[2])
    lista_partidos.append(partido)#Se crea lista de listas

  return lista_partidos

```

```python
def conversion_lista_df (lista_partidos):
  """
  Función que toma la lista de partidos, ya limpios y la convierte en un DF.
  Estableciendo nombres de columnas.

  Argumento recibido: Lista de lista de listas con la info de cada partido.
  Retorna el df de la lista de listas.
  """

  columnas_partidos= ["equipo_local", "puntuacion_equipo_local","puntuacion_equipo_invitado","equipo_invitado"]

  df_partidos = pd.DataFrame(lista_partidos, columns=(columnas_partidos))

  return df_partidos
```

```python
def ampliar_datos_df (df_partidos):
  """
  Función que recoge el df bruto y añade valores por las victorias, empates,
  perdidas y puntos según estas circunstancias. Tanto para el equipo que juega
  en casa como el invitados.

  Argumento recibido: Df con los partidos.
  Retorna df con datos ampliados en los partidos.
  """

  for i, partido in df_partidos.iterrows():
    if partido['puntuacion_equipo_local'] > partido['puntuacion_equipo_invitado']:
      df_partidos.loc[i, 'puntos_local'] = 3
      df_partidos.loc[i, 'local_ganador'] = 1
      df_partidos.loc[i, 'visitante_perdedor'] = 1

    elif partido['puntuacion_equipo_local'] == partido['puntuacion_equipo_invitado']:
      df_partidos.loc[i, 'puntos_local'] = 1
      df_partidos.loc[i, 'puntos_visitante'] = 1
      df_partidos.loc[i, 'local_empate'] = 1
      df_partidos.loc[i, 'visitante_empate'] = 1

    else:
      df_partidos.loc[i, 'puntos_visitante'] = 3
      df_partidos.loc[i, 'local_perdedor'] = 1
      df_partidos.loc[i, 'visitante_ganador'] = 1

  return df_partidos
```

```python
def agrupacion_datos (df_partidos):
  """
  Función que agrupa los resultados y metricas derivadas.
  Devuelve dos  df con los datos de los equipos agrupados, uno con los resultados
  conseguidos jugando en casa y el otro como de invitado.

  Argumento recibido: df partidos con datos ampliados.
  Retorna dos dataframes, agrupando la información por equipo si jugaba en local
  o si jugaba como visitante
  """
  #datos equipo como local:
  df_agrupado_local = df_partidos.groupby('equipo_local')[
    ['puntuacion_equipo_local', #para ver cuantos goles han marcado
     'puntuacion_equipo_invitado', #para ver cuantos goles le han marcado
     'puntos_local',
     'local_ganador',
     'local_empate',
     'local_perdedor']].sum()

  #datos equipo como invitado:
  df_agrupado_invitado = df_partidos.groupby('equipo_invitado')[
    ['puntuacion_equipo_invitado', #para ver cuantos goles han marcado
     'puntuacion_equipo_local', #para ver cuantos goles les han marcado
     'puntos_visitante',
     'visitante_ganador',
     'visitante_empate',
     'visitante_perdedor']].sum()

  return df_agrupado_local, df_agrupado_invitado
```

```python
def suma_datos (df_agrupado_local, df_agrupado_invitado):
  """
  Función que suma los resultados tanto de ida como de vuelta por cada equipo.

  Argumentos recibidos: Dos DataFrames con los datos agrupados por equipo
  uno si jugaba en local y el otro si jugaba como invitado.
  Devuelve un único DF
  """
  df_respuestas = df_agrupado_local['puntuacion_equipo_local'] + df_agrupado_invitado['puntuacion_equipo_invitado'] #el primero sale como serie
  df_respuestas.rename_axis('equipos', inplace=True)
  df_respuestas.rename('goles_a_favor', inplace=True) #se reenombra antes de convertir la serie en DF porque es mas sencillo
  df_respuestas = df_respuestas.to_frame()#convierto la serie en DF
  df_respuestas['goles_en_contra'] = df_agrupado_local['puntuacion_equipo_invitado'] + df_agrupado_invitado['puntuacion_equipo_local']
  df_respuestas['puntos_totales'] = df_agrupado_local['puntos_local'] + df_agrupado_invitado['puntos_visitante']
  df_respuestas['partidos_ganados'] = df_agrupado_local['local_ganador'] + df_agrupado_invitado['visitante_ganador']
  df_respuestas['partidos_empatados'] = df_agrupado_local['local_empate'] + df_agrupado_invitado['visitante_empate']
  df_respuestas['partidos_perdidos'] = df_agrupado_local['local_perdedor'] + df_agrupado_invitado['visitante_perdedor']

  return df_respuestas
```

```python
def resultados (df_respuestas):
  """
  Función que devuelve los resultados a las preguntas solicitadas:
  1. Equipo que más goles a marcado.
  2. Equipo al que más goles han marcado.
  3. Devuelve el DF ordenado por los puntos conseguidos de mayor a menor.

  Argumento recibido: DataFrame con los totales por equipo
  Retorna el DataFrame ordenado por los puntos de mayor a menor
  """
  #1. Devuelve el valor del indice donde se confirma el equipo con la maxima cantidad de goles realizados.
  equipo_mas_goles = df_respuestas['goles_a_favor'].idxmax()
  print("Pregunta 1: Que equipo ha marcado más goles?\n")
  print(f"El equipo que más goles ha marcado es: {equipo_mas_goles}.\n\n")

  #2. Devuelve el valor del indice donde se confirma el equipo con la maxima cantidad de goles recibados.
  equipo_mas_goleado = df_respuestas['goles_en_contra'].idxmax()
  print("Pregunta 2: A qué equipo han marcado más goles?\n")
  print(f"El equipo al que han marcado más goles es: {equipo_mas_goleado}.\n\n")

  #3. Se ordena DF por los puntos totales en orden descendente.
  print("Pregunta 3: Clasificación de los equipos según puntuación total:\n")
  df_final_ordenado =df_respuestas.sort_values(by=['puntos_totales'], ascending=False)
  display(df_final_ordenado)

  return df_final_ordenado
```

```python
def analisis_partidos (ruta):
  """
  Funcion que orquestra todas las funciones necesarias para analizar
  los resultados de los partidos.

  Argumento recibido la ruta del archivo para leer.
  Retorna el DF ordenado por la cantidad de puntos realizados.
  """
  lista_partidos = preparacion_archivo(ruta)
  df_inicio = conversion_lista_df(lista_partidos)
  df_ampliado = ampliar_datos_df(df_inicio)
  df_agrupado_loc, df_agrupado_inv = agrupacion_datos(df_ampliado)
  df_final = suma_datos(df_agrupado_loc, df_agrupado_inv)
  df_final_ordenado = resultados(df_final)

  return df_final_ordenado
```
